```python
highest_rated = data.sort_values(by = ["Star Rating"],ascending = False)
highest_rated = highest_rated.head(10)
print(highest_rated['Product Name'])
```

```python
Lets have a look at the number of ratings of the highest rated i phone on filpkart
```

```python
iphone = highest_rated["Product Name"].value_counts()
labels = iphone.index
counts = highest_rated ["Number Of Ratings"]
figure = px.bar(highest_rated, x=labels, y= counts,
                title ="Number Of Ratings of highest_rated i phones")
figure.show()
```

```python
iphone = highest_rated["Product Name"].value_counts()
labels = iphone.index
counts = highest_rated ["Number Of Reviews"]
figure = px.bar(highest_rated, x=labels, y= counts,
                title ="Number Of Reviews of highest_rated i phones")
figure.show()
```

```python
figure = px.scatter(data_frame = data, x ="Number Of Ratings",
                    y = "Sale Price", size = "Discount Percentage", trendline ="ols",
                    title ="Relationship between sale price and number of rating" )
figure.show()
```

```python
figure = px.scatter(data_frame = data, x ="Number Of Ratings",
                    y = "Discount Percentage", size = "Sale Price", trendline ="ols",
                    title ="Relationship between Discount Percentage and number of rating" )
figure.show()
```

```python
df_filtered_univar_1=data.loc[:,'Product Name':'Sale Price']
df_filtered_univar_1=df_filtered_univar_1.select_dtypes([np.int, np.float])

for i, col in enumerate(df_filtered_univar_1.columns):
    plt.figure(i)
    sns.boxplot(x=col, data=df_filtered_univar_1).set_title("Box Plot of "+col)
```

```python
# Count plot distribution of "BAD" where 0 for loan repaid and 1 for client defaulted on loan
plt.figure(figsize=(12,9))
sns.countplot(x="Sale Price", data=data, palette="Spectral")
plt.title("Plot Distribution of BAD")
plt.show()
```

```python
#ploting the heatmap for correlation
plt.figure(figsize=(18,10))
data=data.corr()
correlation_heatMap = sns.heatmap(data, annot=True)
```
